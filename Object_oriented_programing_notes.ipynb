### 4 PILLARS OR PRINCPLES OF OOP


```python
# 1 ENCAPSULATION
# DEFINITION OF ENCAPSULATION
```

```python
#WITHOUT ENCAPSULATION
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance


account1 = BankAccount("John", 1000)

print("Initial balance:", account1.balance)

# Directly changing balance from outside
account1.balance = -500

print("New balance:", account1.balance)
```

```python
#WITH ENCAPSULATION
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # name mangling

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
        else:
            print("Deposit amount must be positive")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Invalid withdrawal amount")

    def get_balance(self):
        return self.__balance

account = BankAccount(1000)
account.deposit(500)
account.withdraw(200)
print(account.get_balance())
```

### 2. ABSTRACTION
- ABSTRACTION : Abstraction means hiding implementation details and exposing only essential features.

```python
from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def pay(self, amount):
        pass

class CreditCardPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using credit card")

class PayPalPayment(Payment):
    def pay(self, amount):
        print(f"Paid {amount} using PayPal")

payments = [CreditCardPayment(), PayPalPayment()]

for payment in payments:
    payment.pay(100)
```

#IMPORTANCE OF ABTRACTION 
Why abstraction matters
- reduces complexity
- exposes only required behavior
- improves system flexibility

### 3 INHERITANCE
- Inheritance :Inheritance allows one class to acquire properties and methods of another

```python
class Animal:
    def eat(self):
        print("This animal eats food")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

dog = Dog()
dog.eat()
dog.bark()
```

### WHY IT MATTERS
-Benefits
- code reuse
- logical hierarchy
- extension of existing behavior

```python
#SINGLE INHERITANCE
class Vehicle:
    def start(self):
        print("Vehicle started")

class Car(Vehicle):
    def drive(self):
        print("Car is driving")
```

```python
#MULTI-LEVEL INHERITANCE
class LivingBeing:
    def breathe(self):
        print("Breathing")

class Animal(LivingBeing):
    def eat(self):
        print("Eating")

class Dog(Animal):
    def bark(self):
        print("Barking")

dog = Dog()
dog.breathe()
dog.eat()
dog.bark()
```

```python
#HIERACHICAL INHERITANCE
class Employee:
    def work(self):
        print("Employee works")

class Developer(Employee):
    def code(self):
        print("Developer writes code")

class Designer(Employee):
    def design(self):
        print("Designer creates UI")
```

```python
# MULTIPLE INHERITANCE 
class Camera:
    def take_photo(self):
        print("Photo taken")

class Phone:
    def call(self):
        print("Calling...")

class Smartphone(Camera, Phone):
    pass

s = Smartphone()
s.take_photo()
s.call()
```

### POLYMORPHISM
-Polymorphism means one interface, many forms-:Different objects can respond to the same method name in different ways.


```python
#Method overriding example
class Animal:
    def sound(self):
        print("Some generic sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

animals = [Dog(), Cat()]

for animal in animals:
    animal.sound()
```

```python
#POLYMORPHISM WITH FUNCTIONS
class Bird:
    def move(self):
        print("Bird flies")

class Fish:
    def move(self):
        print("Fish swims")

def make_it_move(creature):
    creature.move()

make_it_move(Bird())
make_it_move(Fish())
```

### WHY IT MATTERS 
Why it matters
- makes code extensible
- reduces conditional logic
- supports clean architecture
