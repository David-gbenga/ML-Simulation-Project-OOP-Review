#### RANDOM SOFTWARE ENGINEER QUESTION
Software |Engineer Question Sample 
Given a list of valid customer IDs, a list of valid discount IDs, and a list of discount assignments in the form (customer_id, discount_id), write a validation function that returns False if any customer has been assigned more than 3 discounts. Otherwise return True. You may also validate that all customer IDs and discount IDs in the assignments are valid.

```python
#Random question
def validate_discount_assignments(customers, discounts, discount_assignments):
    valid_customers = set(customers)
    valid_discounts = set(discounts)
    counts = {}

    for customer_id, discount_id in discount_assignments:
        if customer_id not in valid_customers:
            return False

        if discount_id not in valid_discounts:
            return False

        counts[customer_id] = counts.get(customer_id, 0) + 1

        if counts[customer_id] > 3:
            return False

    return True
```

```python
class ZerosOneSolver:
    def parse_shape(self,dims:str)->tuple[int,...]:
        return tuple(map(int,dims.split()))
    def make_zeros(self,shape :tuple[int,...])->np.ndarray:
        return np.zeros(shape, dtype = int)
    def make_ones(self, shape:tuple[int,...])->np.ndarray:
        return np.ones(shape,dtype =int)
    def zero_ones(self,dims:str)->np.ndarray:
        shape = self.parse_shape(dims)
        return self.make_zeros(shape),self.make_ones(shape)
        
u,v =  ZerosOneSolver().zero_ones("2 3")
print(u)
print(v)
```

### 1. Discount Limit Validation 
-Question 
-Given: 
- • a list of valid customer IDs 
- • a list of valid discount IDs 
- • a list of discount assignments in the form (customer_id, discount_id) 
- Write a validation function that returns False if: 
- • any assignment contains an invalid customer ID 
- • any assignment contains an invalid discount ID 
- • any customer has been assigned more than 3 discounts 
- Otherwise return True. 
- Input example 
- customers = [10, 20, 30] 
- discounts = [101, 102, 103, 104] 
- assignments = [(10, 101), (10, 102), (20, 103), (10, 104)] 
- NumPy skills tested 
• np.array 
• slicing columns 
• np.isin 
• np.unique

```python
class DiscountValidator:
    def to_array(self, customers,discounts,assignments)->np.ndarray:
        customers_arr = np.array(customers)
        discounts_arr = np.array(discounts)
        assignments_arr = np.array(assignments)
        return customers_arr,discounts_arr,assignments_arr
    def validate_ids(self,customers_arr,discounts_arr,assignments_arr):
        assigned_customers = assignments_arr[:,0]
        assigned_discounts = assignments_arr[:,1]
        if not np.all(np.isin(assigned_customers,customers_arr)):
            return False
        if not np.all(np.isin(assigned_discounts,discounts_arr)):
            return False 
        return True 
    def validate_discount_limit(self,assignments_arr):
        assigned_customers = assignments_arr[:,0]
        unique_customers, counts = np.unique(assigned_customers,return_counts = True)
        return np.all(counts<=3)
    def solve(self,customers,discounts,assignments):
        customers_arr,discounts_arr ,assignments_arr = self.to_array(customers,discounts,assignments)
        if len(assignments_arr)== 0:
            return True
        if not self.validate_ids(customers_arr,discounts_arr,assignments_arr):
            return False
        return self.validate_discount_limit(assignments_arr)
        
```

```python
#solution : Maximum bets per customer
class MaxBetPlaced:
    #Create array
    def create_array(self,customer_ids, bet_ids, bet_placement):
        customer_arr = np.array(customer_ids)
        bet_ids_arr = np.array(bet_ids)
        bet_placement_arr = np.array(bet_placement)
        return np.array(customer_ids),np.array(bet_ids),np.array(bet_placement)
    #validate customer ids
    def validate_ids(self,customer_arr,bet_ids_arr,bet_placement_arr):
        customer_id_vals = bet_placement_arr[:,0]
        bet_ids_vals =  bet_placement_arr[:,1]
        if not np.all(np.isin(customer_arr,customer_id_vals)):
            return False
        if not np.all(np.isin(bet_ids_arr, bet_ids_vals)):
            return False
        return True
    def max_bet_check(self,bet_placement_arr):
        customer_ids_vals = bet_placement_arr[:,0]
        unique_customer, counts = np.unique(customer_ids_vals,return_counts = True)
        return np.all(counts<=5)
    def max_bet_solver(self,customer_ids, bet_ids, bet_placement):
        customer_arr,bet_ids_arr,bet_placement_arr = self.create_array(customer_ids, bet_ids, bet_placement)
        if not self.validate_ids(customer_arr,bet_ids_arr,bet_placement_arr):
            return False
        return self.max_bet_check(bet_placement_arr)
        
        
```

```python
mport random
import numpy as np


class JaiAlaiGame:
    def __init__(self, players: list[str], average_serve_speed: int = 160, default_rating: int = 1500) -> None:
        self.players = np.array(players)
        self.original_players = np.array(players)
        self.average_serve_speed: int = average_serve_speed
        self.scores = np.zeros(len(players), dtype=int)
        self.winning_score: int = 7
        self.server_idx = 0
        self.challenger_idx = 1
        self.ratings = np.full(len(players), default_rating)
        
#Simulate method to output a tuple with 2 strings
    def simulate_point(self) -> tuple[str, str]:
        #challenger index variable stores a server_idx + 1 value
        challenger_idx = (self.server_idx + 1)
        #The serve_speed variable is defined or is given the sample_serve_speed() method 
        serve_speed = self.sample_serve_speed()
        base_probability_server = self.calculate_base_probability(self.server_idx, challenger_idx)
        win_probability_server = self.adjust_probability_with_serve_speed(base_probability_server, serve_speed)

        winner_idx = self.server_idx if random.random() < win_probability_server else challenger_idx
        loser_idx = challenger_idx if winner_idx == self.server_idx else self.server_idx

        winner_original_idx = np.where(self.original_players == self.players[loser_idx])[0][0]
        self.scores[winner_original_idx] += 1

        if loser_idx == challenger_idx:
            self.players = np.concatenate([[self.players[self.server_idx]], np.roll(self.players[1:], -1)])
        else:
            self.players = np.roll(self.players, -1)

        return self.players[winner_idx], self.players[loser_idx]

    def sample_serve_speed(self) -> int:
        return np.random.poisson(self.average_serve_speed)

    def calculate_base_probability(self, server_idx: int, challenger_idx: int) -> float:
        rating_diff = self.ratings[challenger_idx] - self.ratings[server_idx]
        return 1 / (1 + 10 ** (rating_diff / 400))

    def adjust_probability_with_serve_speed(self, base_probability: float, serve_speed: int) -> float:
        min_speed = 100
        max_speed = 200
        min_adjustment = 0.95
        max_adjustment = 1.05

        serve_speed = np.clip(serve_speed, min_speed, max_speed)
        speed_adjustment = min_adjustment + (serve_speed - min_speed) * (max_adjustment - min_adjustment) / (
                    max_speed - min_speed)
        adjusted_probability = base_probability * speed_adjustment

        return np.clip(adjusted_probability, 0, 1)
        

    def is_game_over(self) -> bool:
        return np.any(self.scores >= self.winning_score)

    def get_winner(self) -> str | None:
        if np.any(self.scores >= self.winning_score):
            return self.original_players[np.argmax(self.scores)]
        return None

    def play_game(self) -> str:
        while not self.is_game_over():
            self.simulate_point()
        return self.get_winner()

    def sim_games(self, simulations: int = 100) -> list:
        winners = []
        for _ in range(simulations):
            self.players = np.array(self.original_players)
            self.scores = np.zeros(len(players), dtype=int)
            winners.append(self.play_game())
        return winners


players = ["Alice", "Bob", "Charlie", "Derek", "Eve", "Frank", "Grace", "Heidi"]
game = JaiAlaiGame(players)
# Get winner
print(game.play_game())

#Calculate the prob of a player winning if we did a thousand simulation
List_players_1000_sim = game.sim_games(1000)
#prob that charlie wins
def prob_charlie_wins(List_players_1000_sim):
    name_to_look_for = "Charlie"
    count = 0 
    for name in List_players_1000_sim:
        if name == name_to_look_for:
            count += 1
    
            
    return count / len(List_players_1000_sim)
    
print(prob_charlie_wins(List_players_1000_sim))
        

```
